Determine the trip time by subtracting `'origin.time_stamp'` from `'destination.time_stamp'`. To get the time in hours, 

```python
from datetime import datetime

# Time elapsed in hours between two timestamps
def time_elapsed(start, end):
  format = '%Y-%m-%dT%H:%M:%S%f%z'
  start = datetime.strptime(start, format)
  end = datetime.strptime(end, format)
  return (end - start).total_seconds() / 3600


# time elapsed in hours between two timestamps in a dataframe
def time_elapsed_df(df, start, end):
    df['trip_time_hours'] = df.apply(lambda x: time_elapsed(x[start], x[end]), axis=1)
    return df

    
time_elapsed_df(knn_df, 'origin.time_stamp', 'destination.time_stamp')
```

```python
def avg_speed(df, distance, time):
  df['avg_speed_kph'] = (df[distance]) / df[time]
  return df

avg_speed(knn_df, 'distance', 'trip_time_hours')
```

```python
# Group timestamps by set periods
def group_timestamps(df, column):
  df['timestamp_year'] = pd.to_datetime(df[column]).dt.year
  df['week_of_year'] = pd.to_datetime(df[column]).dt.isocalendar().week
  df['quarter'] = pd.to_datetime(df[column]).dt.quarter
  df['month'] = pd.to_datetime(df[column]).dt.month_name()
  df['day'] = pd.to_datetime(df[column]).dt.day
  df['day_of_week'] = pd.to_datetime(df[column]).dt.day_name()
  df['day_of_week_id'] = pd.to_datetime(df[column]).dt.dayofweek
  df['hour'] = pd.to_datetime(df[column]).dt.hour
  return df


group_timestamps(knn_df, 'origin.time_stamp')
```
